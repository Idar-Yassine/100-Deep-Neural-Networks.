```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```
xs = np.random.uniform(4, 19, size=(100, 1))
ys = np.sin(xs)
```

```
ax = plt.axes()
ax.scatter(xs, ys, label='Data', edgecolor='darkblue', color='lightblue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right', fancybox=True, shadow=True, frameon=True, framealpha=0.8)
plt.show()
```

```
model = tf.keras.Sequential([
                             tf.keras.layers.Flatten(),
                             tf.keras.layers.Dense(212, activation=tf.nn.relu),
                             tf.keras.layers.Dense(40, activation=tf.nn.sigmoid),
                             tf.keras.layers.Dense(1, activation=tf.nn.tanh)

])
```

```
model.compile(optimizer='adam', loss=tf.keras.losses.mean_squared_error)
```

```
model.fit(xs, ys, epochs=1020)
```

```
ax = plt.axes()
ax.scatter(xs, ys, label='Actual values', edgecolor='darkblue')
ax.scatter(xs, model.predict(xs), label='Predicted values', edgecolor='red', c='darkred')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='upper right', shadow=True, frameon=True, framealpha=0.8)
plt.show()
```
